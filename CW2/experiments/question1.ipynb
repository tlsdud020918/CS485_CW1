```python
import sys
import os
sys.path.append('../')

from scipy import io
import numpy as np
import matplotlib.pyplot as plt
import time
import psutil
from mpl_toolkits.mplot3d import Axes3D
```

```python
vocab_size = [2**n for n in range (1,10)]
data = io.loadmat('.././RF_code/result/question1.mat')

test_histT_list = data['test_histT_list'].flatten()
train_histT_list = data['train_histT_list'].flatten()
vocabT_list = data['vocabT_list'].flatten()

# Display the results
print("vocabT_list:", vocabT_list)
print("train_histT_list:", train_histT_list)
print("test_histT_list:", test_histT_list)
```

```python
# Plotting the data
plt.figure(figsize=(10,6))
plt.plot(vocab_size, test_histT_list, label='train data vector quantization time', marker='o')
plt.plot(vocab_size, train_histT_list, label='test data vector quantization time', marker='s')
plt.plot(vocab_size, vocabT_list, label='K-means clustering time', marker='^')

# Adding labels and title
plt.xlabel('Vocabulary Size K')
plt.ylabel('Elapsed Time (sec)')
plt.title('')
plt.legend()

# Display the plot
plt.savefig('.././report/image/q1-fig1', dpi=300)
```
